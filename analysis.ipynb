```python
df['Product_Rating'] = pd.to_numeric(df['Product_Rating'], errors='coerce')
df['Product_Reviews_Count'] = pd.to_numeric(df['Product_Reviews_Count'], errors='coerce')
df.fillna({'Product_Reviews_Count': 0.0}, inplace=True)
df['Product_Price'] = pd.to_numeric(df['Product_Price'], errors='coerce')

df = df.dropna(subset=['Product_Rating', 'Product_Name'])
df = df.drop_duplicates(subset=['Product_Name', 'Product_Brand'])

print(f"Final rows: {len(df)}")
```

```python
total_products = df['Product_Name'].nunique()
total_brands = df['Product_Brand'].nunique()
total_categories = df['Product_Category'].nunique()
total_reviews = int(df['Product_Reviews_Count'].sum())
avg_price = df['Product_Price'].mean()
meadian_price = df['Product_Price'].median()


print("Total Products :", total_products)
print("Total Brands :", total_brands)
print("Total Categories :", total_categories)
print("Total Reviews :", total_reviews)
print("Average Price :", avg_price)
print("Median Price :", meadian_price)
```

```python
df["Product_Rating"].dtype
```

```python
total_ratings = len(df)
rating_min = df['Product_Rating'].min()
rating_max = df['Product_Rating'].max()
avg_rating = df['Product_Rating'].mean()


print("Total Rating : ", total_ratings)
print("Minimum Rating : ", rating_min)
print("Maximum Rating : ", rating_max)
print("Average Rating : ", avg_rating)
```

```python
product_stats = df.groupby('Product_Name').agg({
    'Product_Rating': 'mean',
    'Product_Reviews_Count': 'sum',
    'Product_Price': 'mean',
    'Product_Brand': 'first',
    'Product_Category': 'first'
}).round(2)

product_stats.columns = ['avg_rating', 'total_reviews', 'avg_price', 'brand', 'category'] 
product_stats = product_stats.reset_index()

product_stats = product_stats.sort_values('avg_rating', ascending=False)


print("Top 10 highest Rated Products : ")
print(product_stats.head(10)[['Product_Name', 'avg_rating', 'total_reviews', 'avg_price', 'brand', 'category']])


print("Top 10 most reviewed : ")
print(product_stats.nlargest(10, 'total_reviews')[['Product_Name', 'avg_rating', 'total_reviews', 'avg_price', 'brand', 'category']])
```

```python
# product with high ratings like >4.5 but low reviews < 10

hidden_gems = product_stats[
    (product_stats['avg_rating'] >= 4.5) &
    (product_stats['total_reviews'] < 10)
]

print(f"\nFound {len(hidden_gems)} hidden gem products")
print(f"These represent {len(hidden_gems)/len(product_stats)*100:.1f}% of all products")
print("\nTop Hidden Gems:")
print(hidden_gems.head(10)[['Product_Name', 'brand', 'avg_rating', 'total_reviews']])

```

```python
category_stats = df.groupby('Product_Category').agg({
    'Product_Rating': 'mean',
    'Product_Reviews_Count': 'sum',
    'Product_Brand' : 'first',
    'Product_Price': 'mean',
    'Product_Name': 'count'
}).round(2)

category_stats.columns = ['avg_rating', 'total_reviews','brand', 'avg_price', 'product_count']
category_stats = category_stats.reset_index()
category_stats = category_stats.sort_values('avg_rating', ascending=False)

print(f"\nCategory Performance:")
print(category_stats)
```

```python
best_category = category_stats.iloc[0]
worst_category = category_stats.iloc[-1]

print(f"\nBest Performing Category: {best_category['Product_Category']}")
print(f"  - Avg Rating: {best_category['avg_rating']:.2f}")
print(f"  - Products: {int(best_category['product_count'])}")
print(f"  - Brands: {(best_category['brand'])}")

print(f"\nWorst Performing Category: {worst_category['Product_Category']}")
print(f"  - Avg Rating: {worst_category['avg_rating']:.2f}")
print(f"  - Products: {int(worst_category['product_count'])}")
print(f"  - Brands: {(worst_category['brand'])}")

```

```python
brand_stats = df.groupby('Product_Brand').agg({
    'Product_Rating': 'mean',
    'Product_Reviews_Count': 'sum',
    'Product_Brand' : 'first',
    'Product_Price': 'mean',
    'Product_Name': 'count'
}).round(2)

brand_stats.columns = ['avg_rating', 'total_reviews', 'brand', 'avg_price', 'product_count']
brand_stats = brand_stats.reset_index()

# Top brands by rating (minimum 5 products)
top_brands = brand_stats[brand_stats['product_count'] >= 5].nlargest(10, 'avg_rating')

print(f"\nTop 10 Brands by Rating (min 5 products):")
print(top_brands[['Product_Brand', 'avg_rating', 'product_count', 'total_reviews']])
```

```python
rating_dist = df['Product_Rating'].value_counts().sort_index()
print(f"\nRating Distribution:")
print(rating_dist)

# Percentage breakdown
positive_reviews = len(df[df['Product_Rating'] >= 4.0])
neutral_reviews = len(df[(df['Product_Rating'] >= 3.0) & (df['Product_Rating'] < 4.0)])
negative_reviews = len(df[df['Product_Rating'] < 3.0])

total = len(df)
print(f"\nRating Breakdown:")
print(f"Positive (4.0-5.0): {positive_reviews:,} ({positive_reviews/total*100:.1f}%)")
print(f"Neutral (3.0-3.9): {neutral_reviews:,} ({neutral_reviews/total*100:.1f}%)")
print(f"Negative (<3.0): {negative_reviews:,} ({negative_reviews/total*100:.1f}%)")
```

```python
print(f"\nPrice Statistics:")
print(df['Product_Price'].describe())

# Price vs Rating correlation
price_rating_corr = df['Product_Price'].corr(df['Product_Rating'])
print(f"\nCorrelation between Price and Rating: {price_rating_corr:.3f}")

# Price segments
df['price_segment'] = pd.cut(df['Product_Price'], 
                              bins=[0, 500, 1000, 2000, float('inf')],
                              labels=['Budget (<₹500)', 'Mid (₹500-1K)', 'Premium (₹1K-2K)', 'Luxury (>₹2K)'])

price_segment_stats = df.groupby('price_segment').agg({
    'Product_Rating': 'mean',
    'Product_Reviews_Count': 'mean',
    'Product_Name': 'count'
}).round(2)

print(f"\nPerformance by Price Segment:")
print(price_segment_stats)
```

```python
df.to_csv('cleaned_nykaa_data.csv', index=False)
print("✓ cleaned_nykaa_data.csv")

# Product summary
product_stats.to_csv('product_summary.csv', index=False)
print("✓ product_summary.csv")

# Category summary
category_stats.to_csv('category_summary.csv', index=False)
print("✓ category_summary.csv")

# Brand summary
brand_stats.to_csv('brand_summary.csv', index=False)
print("✓ brand_summary.csv")

# Hidden gems
hidden_gems.to_csv('hidden_gems.csv', index=False)
print("✓ hidden_gems.csv")

print("\nAll files ready for Power BI!")
```
